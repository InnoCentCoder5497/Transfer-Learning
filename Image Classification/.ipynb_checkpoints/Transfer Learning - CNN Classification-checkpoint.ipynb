# Transfer Learning


## Import Libraries

```python
# basic manupulation
import numpy as np
# Plotting
import seaborn
import matplotlib.pyplot as plt
# PyTorch for neural network
import torch
import torch.nn as nn
import torch.optim as opt
from torch.utils.data import DataLoader
# torchvision for image manupulations and pretrained model
from torchvision import datasets
from torchvision import transforms, utils
from torchvision import models
# setting manual seed for reproducibility
torch.manual_seed(101)
```

```python
# Defining data transforms, our pretrained model requires the data to be 
# in the shape (3, 224, 224) in (channel, width, height) format
# so we first resize the images to that size, then apply a horizontal flip
# randomly with probability 0.5. Then we scale the pixel values in the range
# [0-1]. After that we normalise the image pixels to have a mean of 
# [0.485, 0.456, 0.406] in each channel and have a standard deviation of
# [0.229, 0.224, 0.225] in each channel.

data_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Creating a pytorch dataset from the image folder
data = datasets.ImageFolder('train/', transform=data_transform)

# Applying sample steps of transforms as in training data except for the 
# Random flip as data augmentation is not required
test_data_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Creating pytorch dataset for test set
test_data = datasets.ImageFolder('test/', transform=test_data_transform)

# dictionary to convert indexes to class names
idx_to_class = dict(zip(data.class_to_idx.values(), data.class_to_idx.keys()))
# data loaders for training and testing
dloader = DataLoader(data, batch_size = 32, shuffle=True)
test_loader = DataLoader(test_data, batch_size = 32, shuffle=True)
```

```python
# function to show images from a batch
def show_images(batch, num_images = 5, prediction = None):
    """
    arguments:
        - batch : tuple, form (images tensor, labels tensor)
        - num_images : int, number of images to display
        - predictions : tensor, if given display the predicted class name also
        
    output:
        - plots the images from the given batch
    """
    allimgs = batch[0].numpy()
    labels = batch[1].numpy()
    
    if prediction is not None:
        prediction = prediction.numpy()
    
    fig = plt.figure(figsize = (10, 10))
    
    for i in range(num_images):
        img = allimgs[i, :, :, :]
        img = img.transpose((1, 2, 0))
        img = (img * np.array([0.229, 0.224, 0.225])) + np.array([0.485, 0.456, 0.406])
        img = np.clip(img, 0, 1)
        fig.add_subplot(1, num_images, i+1)
        if prediction is not None:
            plt.title("Original : {}\nPredicted : {}".format(idx_to_class[labels[i]], idx_to_class[prediction[i]]))
        else:
            plt.title(idx_to_class[labels[i]])
        plt.imshow(img)
```

```python
# getting a sample batch
img_batch = next(iter(dloader))

# showing 5 images from the sampled batch
show_images(img_batch)
```

## Model as Feature extractor

```python
# number of classes in our dataset
num_classes = len(data.classes)
# loading pretrained network
resnet = models.resnet18(pretrained=True)
# setting require_grad to False, this freezes all layers,
# which makes pytorch not to calculate gradients for the layers
# keeping them un-altered during training process 
for param in resnet.parameters():
    param.require_grad = False
# number of features output by the pretrained model to the last layer
feature_out = resnet.fc.in_features
# new custom last layer. NOTE: new layers have the require_grad parameter 
# set to True by default
last_layer = nn.Linear(feature_out, num_classes)
# replacing last layer of pretrained network
resnet.fc = last_layer
# sending the network to gpu
resnet = resnet.cuda()
```

```python
# Loss function
cost = nn.CrossEntropyLoss()
# optimizer and its parameters
LR = 0.0001
adam = opt.Adam(resnet.parameters(), lr = LR)
```

```python
# function to run test set
def run_test_set(model, data_loader):
    """
    arguments:
        - model : pytorch model to test
        - data_loader : testing data loader to test on
        
    returns:
        - accuracy : float, accuracy on testing set
    """
    # set model in testing mode
    model = model.eval()
    # book keeping
    t_correct = 0
    total_samples = 0
    # run testing
    for batch in data_loader:
        images = batch[0].cuda()
        labels = batch[1]

        pred_logits = model(images)

        _, pred = torch.max(pred_logits, 1)

        pred = pred.cpu()
        # calculate correct predictions in mini-batch
        correct = (pred == labels).sum()
        t_correct += correct.item()
        total_samples += len(labels)

    return t_correct / total_samples
```

```python
N_EPOCHS = 5
# training loop
# set model in training mode
resnet = resnet.train()
for epoch in range(N_EPOCHS):
    # book keeping
    epoch_loss = []
    for i, batch in enumerate(dloader):
        # reset all gradients
        adam.zero_grad()
        # send input and target to gpu
        images = batch[0].cuda()
        labels = batch[1].cuda()
        # run forward pass
        output = resnet(images)
        # calculate loss and store it
        loss = cost(output, labels)
        epoch_loss.append(loss.item())
        # calculate gradients
        loss.backward()
        # run optimiser
        adam.step()
    
    # run network on testing set after each epoch
    test_acc = run_test_set(resnet, test_loader)
    print("Epoch : {}, loss = {:.5f}, test Accuracy : {:.4f}%".format(epoch + 1, sum(epoch_loss) / len(epoch_loss), test_acc * 100))
```

```python
# function to get predictions from model
def get_prediction(model, images):
    """
    arguments:
        - model : pytorch model to run
        - images : tensor, images to get predictions of
        
    returns:
        - predictions : tensor containing indexes of predicted class
    """
    # model in evaluation mode
    model = model.eval()
    images = images.cuda()

    pred_logits = model(images)
    _, pred = torch.max(pred_logits, 1)

    return pred.cpu()
```

```python
# generate a batch from test set
sample = next(iter(test_loader))

# get first 5 images and labels
imgs = sample[0][:5, :, :, :]
labels = sample[1][:5]

# run model
predictions = get_prediction(resnet, imgs)
# display images with predictions
display_batch = (imgs, labels)
show_images(display_batch, prediction=predictions)
```

## Fine Tuning the Whole model
The only difference from above is that, we finetune the whole model, instead of just the final layer we are adding. So, we will not set the `require_grad` parameter of later to `False` while finetuning the whole network.

```python
resnet = models.resnet18(pretrained=True)
feature_out = resnet.fc.in_features

last_layer = nn.Linear(feature_out, num_classes)
resnet.fc = last_layer

resnet = resnet.cuda()
```

```python
cost = nn.CrossEntropyLoss()

adam = opt.Adam(resnet.parameters(), lr = LR)
```

```python
for epoch in range(N_EPOCHS):
    epoch_loss = []
    for i, batch in enumerate(dloader):
        adam.zero_grad()

        images = batch[0].cuda()
        labels = batch[1].cuda()

        output = resnet(images)

        loss = cost(output, labels)
        epoch_loss.append(loss.item())
        loss.backward()

        adam.step()
        
    test_acc = run_test_set(resnet, test_loader)
    print("Epoch : {}, loss = {:.5f}, test Accuracy : {:.4f}%".format(epoch + 1, sum(epoch_loss) / len(epoch_loss), test_acc * 100))
```

```python
sample = next(iter(test_loader))

imgs = sample[0][:5, :, :, :]
labels = sample[1][:5]

display_batch = (imgs, labels)
predictions = get_prediction(resnet, imgs)

show_images(display_batch, prediction=predictions)
```
